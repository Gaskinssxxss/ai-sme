```python
independent_vars = ['Umur', 'Jenis Kelamin', 'Pekerjaan', 'Pendapatan', 'Pendidikan', 'Status Kawin', 'Kecamatan', 'Kelurahan']
dependent_var = '12. Dari daftar nama Kandidat yang telah beredar yang akan maju pada pilkada mendatang. Jika Pilkada dilaksanakan sekarang, dari nama-nama yang Bapak/Ibu/Saudara kenal atau tahu, manakah yang akan anda pilih sebagai Walikota Bima?'

file_path = 'DATA.xlsx'
df = pd.read_excel(file_path, engine='openpyxl')

bootstrap_results = {}

for var in independent_vars:
    filtered_data = df[[var, dependent_var]].dropna()
    resampled_data = resample(filtered_data, replace=True, n_samples=8100, random_state=42)
    bootstrap_results[var] = resampled_data

output_dir = "bootstrap_results"
os.makedirs(output_dir, exist_ok=True)

for var, data in bootstrap_results.items():
    output_path = os.path.join(output_dir, f"bootstrap_{var}.xlsx")
    data.to_excel(output_path, index=False)
    print(f"Resampled data for {var} saved to {output_path}")
    
print("\nAll resampled datasets have been saved as Excel files.")
```

```python
input_dir = "bootstrap_results"
files = os.listdir(input_dir)

target_column = '12. Dari daftar nama Kandidat yang telah beredar yang akan maju pada pilkada mendatang. Jika Pilkada dilaksanakan sekarang, dari nama-nama yang Bapak/Ibu/Saudara kenal atau tahu, manakah yang akan anda pilih sebagai Walikota Bima?'

output_dir = "knn_results"
os.makedirs(output_dir, exist_ok=True)

for file in files:
    file_path = os.path.join(input_dir, file)
    
    data = pd.read_excel(file_path)
    
    feature_column = data.columns[0]
    X = data[[feature_column]]
    y = data[target_column]
    
    if X[feature_column].dtype == 'object':
        X = pd.get_dummies(X, drop_first=True)
    if y.dtype == 'object':
        y = pd.factorize(y)[0]
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
    
    knn = KNeighborsClassifier(n_neighbors=5)
    knn.fit(X_train, y_train)
    
    y_pred = knn.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, output_dict=True, zero_division=1)
    
    table = PrettyTable()
    table.field_names = ["Metric", "Kandidat 0", "Kandidat 1", "Kandidat 2", "Weighted Avg"]
    
    for metric in ["precision", "recall", "f1-score"]:
        table.add_row([
            metric.capitalize(),
            f"{report['0'][metric]:.2f}" if '0' in report else "N/A",
            f"{report['1'][metric]:.2f}" if '1' in report else "N/A",
            f"{report['2'][metric]:.2f}" if '2' in report else "N/A",
            f"{report['weighted avg'][metric]:.2f}" if 'weighted avg' in report else "N/A",
        ])
    
    output_file = os.path.join(output_dir, f"{feature_column}.txt")
    try:
        with open(output_file, "w") as f:
            f.write(f"Results for Feature: {feature_column} (File: {file})\n")
            f.write(f"Accuracy: {accuracy:.2f}\n")
            f.write("\nDetailed Classification Metrics:\n")
            f.write(table.get_string())
            if any(report[str(cls)]["precision"] == 0 for cls in range(len(set(y_test)))):
                f.write("\n\nWarning: Some classes have no predictions. Consider addressing class imbalance.\n")
            print(f"{feature_column} : 200 OK!")
    except:
        print("500 upps!")
    
    if any(report[str(cls)]["precision"] == 0 for cls in range(len(set(y_test)))):
        print("\nWarning: Some classes have no predictions. Consider addressing class imbalance.")
    
        summary_table.add_row([
        feature_column,
        f"{accuracy:.2f}",
        f"{report['weighted avg']['precision']:.2f}",
        f"{report['weighted avg']['recall']:.2f}",
        f"{report['weighted avg']['f1-score']:.2f}"
    ])

summary_output_file = os.path.join(output_dir, "summary.txt")
try:
    with open(summary_output_file, "w") as f:
        f.write("Summary of Results for All Features:\n")
        f.write(summary_table.get_string())
        print("Summary : 200 OK!")
except:
    print("500 uppps!")
```
